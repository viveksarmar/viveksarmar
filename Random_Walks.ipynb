```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
years = 20  # Number of years to simulate
initial_price = 100  # Initial stock price
num_simulations = 5  # Number of simulations to run

# Upward and downward movement parameters
up_mu, up_sigma = 0.20, 0.10  # Mean and standard deviation for upward movements
down_mu, down_sigma = -0.10, 0.05  # Mean and standard deviation for downward movements

# Time array
time = np.arange(years)

# Initialize stock price array
stock_prices = np.zeros(years)
stock_prices[0] = initial_price

# Simulate stock price movements
for t in range(1, years):
    if t % 2 == 1:  # Upward movement (odd years)
        movement = np.random.normal(up_mu, up_sigma)
    else:  # Downward movement (even years)
        movement = np.random.normal(down_mu, down_sigma)

    stock_prices[t] = stock_prices[t-1] * (1 + movement)

# Plot the stock price over time
plt.figure(figsize=(10, 6))
plt.plot(time, stock_prices, marker='o', linestyle='-', color='b')
plt.title('Simulated Stock Price with Upward and Downward Movements')
plt.xlabel('Years')
plt.ylabel('Stock Price')
plt.grid(True)
plt.show()
```
